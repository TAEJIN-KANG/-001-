```python
import pandas as pd
import numpy as np

# CSV 파일 경로
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'vehicle_prod.csv'

# 데이터프레임 읽기
df = pd.read_csv(file, index_col=0)

# 숫자형 데이터의 합계 계산
sums = df.select_dtypes(np.number).sum().rename('Total')
means = df.select_dtypes(np.number).mean().rename('Average')

# 합계 데이터프레임 생성
sums_df = pd.DataFrame(sums).T
means_df = pd.DataFrame(means).T

sums_df.index = ['Total']
means_df.index = ['Average']

# 원본 데이터프레임과 합계 데이터프레임 결합
total_df = pd.concat([df, sums_df,means_df], ignore_index=False)

# 결과 출력
total_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')
weather['month'] = pd.DatetimeIndex(weather['일시']).month

monthly = [None for x in range(12)]
monthly_temp = [0 for x in range(12)]
monthly_wind = [0 for x in range(12)]

for i in range(12):
    monthly[i] = weather[weather['month'] == i+1]
    #monthly_temp[i] = monthly[i]['평균기온'].mean()
    monthly_wind[i] = monthly[i]['평균풍속'].mean()

months = np.arange(1,13)
#plt.bar(months, monthly_temp, color='green')
plt.bar(months, monthly_wind, color='red')
plt.xlabel('month')
plt.ylabel('speed')
plt.show()
```
