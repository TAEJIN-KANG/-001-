```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

m = 50  #100에서 50포인트로 변경
x = 8 * np.random.rand(m, 1) - 4
y = 0.5 * x**2 + 0.5 * x + 1 + np.random.randn(m, 1)

poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x)
regr = LinearRegression()
regr.fit(x_poly, y)

domain = np.linspace(-4, 4, 50).reshape(-1, 1)
domain2 = poly_features.fit_transform(domain)
y_predict = regr.predict(domain2)

plt.figure(figsize=(5, 3))  #그래프 사이즈(5,3)inch로 변경
plt.scatter(x, y)
plt.plot(domain, y_predict, color='yellow', linewidth=3)  # 색상 노랑,선굵기3으로 변경
plt.show()
```
